```python
from sklearn.model_selection import train_test_split
import torch.utils.data as data

torch.manual_seed(42)
np.random.seed(42)

def create_data_loaders(dataset, batch_size=8, train_split=0.8):
    train_size = int(train_split * len(dataset))
    val_size = len(dataset) - train_size

    train_dataset, val_dataset = data.random_split(
        dataset,
        [train_size, val_size]
    )

    train_loader = data.DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=2
    )

    val_loader = data.DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=2
    )

    return train_loader, val_loader

dataset = LiverMRIDataset(BASE_DIR, transform=train_transforms)

train_loader, val_loader = create_data_loaders(dataset)

print(f"Total dataset size: {len(dataset)}")
print(f"Training samples: {len(train_loader.dataset)}")
print(f"Validation samples: {len(val_loader.dataset)}")

for images, labels in train_loader:
    print(f"Batch shape: {images.shape}")
    print(f"Labels shape: {labels.shape}")
    break
```

```python
import torch.nn as nn
from torchvision.models import resnet50, ResNet50_Weights

class HCCModel(nn.Module):
    def __init__(self):
        super(HCCModel, self).__init__()
        self.resnet = resnet50(weights=ResNet50_Weights.DEFAULT)

        self.resnet.conv1 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)

        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.resnet(x)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = HCCModel().to(device)
print(f"Using device: {device}")

criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    best_val_loss = float('inf')
    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': []
    }

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.float().to(device)

            optimizer.zero_grad()

            outputs = model(inputs).squeeze()
            loss = criterion(outputs, labels)

            loss.backward()
            optimizer.step()

            predictions = (outputs > 0.5).float()
            train_correct += (predictions == labels).sum().item()
            train_total += labels.size(0)
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.float().to(device)
                outputs = model(inputs).squeeze()
                loss = criterion(outputs, labels)

                predictions = (outputs > 0.5).float()
                val_correct += (predictions == labels).sum().item()
                val_total += labels.size(0)
                val_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)
        train_acc = train_correct / train_total * 100
        val_acc = val_correct / val_total * 100

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        print(f'Epoch [{epoch+1}/{num_epochs}]:')
        print(f'Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print('-' * 50)

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), 'best_model.pth')

    return history

history = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10)
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau

class EarlyStopping:
    def __init__(self, patience=5, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=20):
    early_stopping = EarlyStopping(patience=5)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)
    best_val_loss = float('inf')
    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': []
    }

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.float().to(device)
            optimizer.zero_grad()
            outputs = model(inputs).squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            predictions = (outputs > 0.5).float()
            train_correct += (predictions == labels).sum().item()
            train_total += labels.size(0)
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.float().to(device)
                outputs = model(inputs).squeeze()
                loss = criterion(outputs, labels)

                predictions = (outputs > 0.5).float()
                val_correct += (predictions == labels).sum().item()
                val_total += labels.size(0)
                val_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)
        train_acc = train_correct / train_total * 100
        val_acc = val_correct / val_total * 100

        scheduler.step(avg_val_loss)

        early_stopping(avg_val_loss)


        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        print(f'Epoch [{epoch+1}/{num_epochs}]:')
        print(f'Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print('-' * 50)

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), 'best_model.pth')

        if early_stopping.early_stop:
            print("Early stopping triggered")
            break

    return history

model = HCCModel().to(device)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=0.001)

history = train_model(model, train_loader, val_loader, criterion, optimizer)
```

```python
best_model = HCCModel().to(device)
best_model.load_state_dict(torch.load('best_model.pth'))

def evaluate_model(model, data_loader):
    model.eval()
    correct = 0
    total = 0
    predictions_list = []
    true_labels_list = []

    with torch.no_grad():
        for inputs, labels in data_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs).squeeze()
            predictions = (outputs > 0.5).float()

            correct += (predictions == labels).sum().item()
            total += labels.size(0)

            predictions_list.extend(predictions.cpu().numpy())
            true_labels_list.extend(labels.cpu().numpy())

    accuracy = 100 * correct / total
    return accuracy, predictions_list, true_labels_list

train_acc, train_preds, train_labels = evaluate_model(best_model, train_loader)
val_acc, val_preds, val_labels = evaluate_model(best_model, val_loader)

print(f"Final Train Accuracy: {train_acc:.2f}%")
print(f"Final Validation Accuracy: {val_acc:.2f}%")
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import nibabel as nib
from torchvision import transforms, models
from torchvision.models import resnet50, ResNet50_Weights
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve, auc

device = torch.device('cpu')
print(f"Using device: {device}")

train_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.1, contrast=0.1),
])

class LiverMRIDataset(Dataset):
    def __init__(self, base_dir, transform=None, train=True):
        self.base_dir = base_dir
        self.transform = transform
        self.train = train
        self.samples = []

        hcc_dir = os.path.join(base_dir, 'HCC')
        for patient_id in os.listdir(hcc_dir):
            if patient_id != '.DS_Store':
                patient_path = os.path.join(hcc_dir, patient_id)
                art_path = os.path.join(patient_path, 'art.nii.gz')
                pv_path = os.path.join(patient_path, 'pv.nii.gz')
                del_path = os.path.join(patient_path, 'del.nii.gz')
                if os.path.exists(art_path) and os.path.exists(pv_path) and os.path.exists(del_path):
                    self.samples.append((art_path, pv_path, del_path, 1))

        normal_dir = os.path.join(base_dir, 'Normal')
        for patient_id in os.listdir(normal_dir):
            if patient_id != '.DS_Store':
                patient_path = os.path.join(normal_dir, patient_id)
                art_path = os.path.join(patient_path, 'art.nii.gz')
                pv_path = os.path.join(patient_path, 'pv.nii.gz')
                del_path = os.path.join(patient_path, 'del.nii.gz')
                if os.path.exists(art_path) and os.path.exists(pv_path) and os.path.exists(del_path):
                    self.samples.append((art_path, pv_path, del_path, 0))

    def preprocess_volume(self, image):
        image = (image - np.mean(image)) / np.std(image)
        image = np.clip(image, -3, 3)
        image = (image + 3) / 6
        mid_slice = image[:, :, image.shape[2]//2]
        mid_slice = transforms.functional.resize(
            torch.FloatTensor(mid_slice).unsqueeze(0),
            (224, 224)
        ).squeeze(0)
        return mid_slice

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        art_path, pv_path, del_path, label = self.samples[idx]
        art_img = self.preprocess_volume(nib.load(art_path).get_fdata())
        pv_img = self.preprocess_volume(nib.load(pv_path).get_fdata())
        del_img = self.preprocess_volume(nib.load(del_path).get_fdata())
        image = torch.stack([art_img, pv_img, del_img])
        if self.train and self.transform:
            image = self.transform(image)
        return image, label

class HCCModel(nn.Module):
    def __init__(self):
        super(HCCModel, self).__init__()
        self.resnet = resnet50(weights=ResNet50_Weights.DEFAULT)
        self.resnet.conv1 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.resnet(x)

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    best_val_loss = float('inf')
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.float().to(device)
            optimizer.zero_grad()
            outputs = model(inputs).squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            predictions = (outputs > 0.5).float()
            train_correct += (predictions == labels).sum().item()
            train_total += labels.size(0)
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.float().to(device)
                outputs = model(inputs).squeeze()
                loss = criterion(outputs, labels)

                predictions = (outputs > 0.5).float()
                val_correct += (predictions == labels).sum().item()
                val_total += labels.size(0)
                val_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)
        train_acc = train_correct / train_total * 100
        val_acc = val_correct / val_total * 100

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        print(f'Epoch [{epoch+1}/{num_epochs}]:')
        print(f'Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print('-' * 50)

    return history

def k_fold_cross_validation(dataset, k=5, num_epochs=10):
    kfold = KFold(n_splits=k, shuffle=True, random_state=42)
    fold_results = []
    indices = np.arange(len(dataset))

    for fold, (train_idx, val_idx) in enumerate(kfold.split(indices)):
        print(f"\nFold {fold+1}/{k}")
        print("-" * 50)

        train_sampler = torch.utils.data.SubsetRandomSampler(train_idx)
        val_sampler = torch.utils.data.SubsetRandomSampler(val_idx)

        train_loader = torch.utils.data.DataLoader(
            dataset, batch_size=4, sampler=train_sampler, num_workers=0
        )
        val_loader = torch.utils.data.DataLoader(
            dataset, batch_size=4, sampler=val_sampler, num_workers=0
        )

        model = HCCModel().to(device)
        criterion = nn.BCELoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=0.001)

        history = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs)

        model.eval()
        val_probs = []
        val_labels_list = []
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs = inputs.to(device)
                outputs = model(inputs).squeeze()
                val_probs.extend(outputs.cpu().numpy())
                val_labels_list.extend(labels.numpy())

        fpr, tpr, _ = roc_curve(val_labels_list, val_probs)
        roc_auc = auc(fpr, tpr)

        fold_results.append({
            'auc': roc_auc,
            'history': history
        })

        print(f"Fold {fold+1} - AUC: {roc_auc:.3f}")

    avg_auc = np.mean([r['auc'] for r in fold_results])
    print(f"\nAverage AUC across folds: {avg_auc:.3f}")

    return fold_results

if __name__ == "__main__":
    BASE_DIR = '/content/drive/MyDrive/Dataset'
    dataset = LiverMRIDataset(BASE_DIR, transform=train_transforms)
    results = k_fold_cross_validation(dataset, k=5)
```

```python
from IPython.display import display, Javascript
display(Javascript('''
function ClickConnect(){
    console.log("Working");
    document.querySelector("#top-toolbar > colab-connect-button").shadowRoot.querySelector("#connect").click()
}
setInterval(ClickConnect, 60000)
'''))

import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import nibabel as nib
from torchvision import transforms, models
from torchvision.models import resnet50, ResNet50_Weights
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve, auc

device = torch.device('cpu')
print(f"Using device: {device}")

train_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1))
])

class LiverMRIDataset(Dataset):
    def __init__(self, base_dir, transform=None, train=True):
        self.base_dir = base_dir
        self.transform = transform
        self.train = train
        self.samples = []

        hcc_dir = os.path.join(base_dir, 'HCC')
        for patient_id in os.listdir(hcc_dir):
            if patient_id != '.DS_Store':
                patient_path = os.path.join(hcc_dir, patient_id)
                art_path = os.path.join(patient_path, 'art.nii.gz')
                pv_path = os.path.join(patient_path, 'pv.nii.gz')
                del_path = os.path.join(patient_path, 'del.nii.gz')
                if os.path.exists(art_path) and os.path.exists(pv_path) and os.path.exists(del_path):
                    self.samples.append((art_path, pv_path, del_path, 1))

        normal_dir = os.path.join(base_dir, 'Normal')
        for patient_id in os.listdir(normal_dir):
            if patient_id != '.DS_Store':
                patient_path = os.path.join(normal_dir, patient_id)
                art_path = os.path.join(patient_path, 'art.nii.gz')
                pv_path = os.path.join(patient_path, 'pv.nii.gz')
                del_path = os.path.join(patient_path, 'del.nii.gz')
                if os.path.exists(art_path) and os.path.exists(pv_path) and os.path.exists(del_path):
                    self.samples.append((art_path, pv_path, del_path, 0))

    def preprocess_volume(self, image):
        image = (image - np.mean(image)) / np.std(image)
        image = np.clip(image, -3, 3)
        image = (image + 3) / 6
        mid_slice = image[:, :, image.shape[2]//2]
        mid_slice = transforms.functional.resize(
            torch.FloatTensor(mid_slice).unsqueeze(0),
            (224, 224)
        ).squeeze(0)
        return mid_slice

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        art_path, pv_path, del_path, label = self.samples[idx]
        art_img = self.preprocess_volume(nib.load(art_path).get_fdata())
        pv_img = self.preprocess_volume(nib.load(pv_path).get_fdata())
        del_img = self.preprocess_volume(nib.load(del_path).get_fdata())
        image = torch.stack([art_img, pv_img, del_img])
        if self.train and self.transform:
            image = self.transform(image)
        return image, label

class HCCModel(nn.Module):
    def __init__(self, dropout_rate=0.5):
        super(HCCModel, self).__init__()
        self.resnet = resnet50(weights=ResNet50_Weights.DEFAULT)
        self.resnet.conv1 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.resnet(x)

def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=10):
    best_val_loss = float('inf')
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.float().to(device)
            optimizer.zero_grad()
            outputs = model(inputs).squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            predictions = (outputs > 0.5).float()
            train_correct += (predictions == labels).sum().item()
            train_total += labels.size(0)
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.float().to(device)
                outputs = model(inputs).squeeze()
                loss = criterion(outputs, labels)

                predictions = (outputs > 0.5).float()
                val_correct += (predictions == labels).sum().item()
                val_total += labels.size(0)
                val_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)
        train_acc = train_correct / train_total * 100
        val_acc = val_correct / val_total * 100

        scheduler.step(avg_val_loss)

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        print(f'Epoch [{epoch+1}/{num_epochs}]:')
        print(f'Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print('-' * 50)

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), f'best_model_fold.pth')

    return history

def k_fold_cross_validation(dataset, k=5, num_epochs=10):
    kfold = KFold(n_splits=k, shuffle=True, random_state=42)
    fold_results = []
    indices = np.arange(len(dataset))

    for fold, (train_idx, val_idx) in enumerate(kfold.split(indices)):
        print(f"\nFold {fold+1}/{k}")
        print("-" * 50)

        train_sampler = torch.utils.data.SubsetRandomSampler(train_idx)
        val_sampler = torch.utils.data.SubsetRandomSampler(val_idx)

        train_loader = torch.utils.data.DataLoader(
            dataset, batch_size=4, sampler=train_sampler, num_workers=0
        )
        val_loader = torch.utils.data.DataLoader(
            dataset, batch_size=4, sampler=val_sampler, num_workers=0
        )

        model = HCCModel(dropout_rate=0.5).to(device)
        criterion = nn.BCELoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=0.00005, weight_decay=0.01)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=3, verbose=True
        )

        history = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs)

        model.eval()
        val_probs = []
        val_labels_list = []
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs = inputs.to(device)
                outputs = model(inputs).squeeze()
                val_probs.extend(outputs.cpu().numpy())
                val_labels_list.extend(labels.numpy())

        fpr, tpr, _ = roc_curve(val_labels_list, val_probs)
        roc_auc = auc(fpr, tpr)

        fold_results.append({
            'auc': roc_auc,
            'history': history
        })

        print(f"Fold {fold+1} - AUC: {roc_auc:.3f}")

    avg_auc = np.mean([r['auc'] for r in fold_results])
    print(f"\nAverage AUC across folds: {avg_auc:.3f}")

    return fold_results

if __name__ == "__main__":
    BASE_DIR = '/content/drive/MyDrive/Dataset'
    dataset = LiverMRIDataset(BASE_DIR, transform=train_transforms)
    results = k_fold_cross_validation(dataset, k=5)
```
